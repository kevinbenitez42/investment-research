```python
#plot broad_market data on same chart which is a dictionary , key is name, value is a dataframes with columns 'Close', 'Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits'
def plot_market_data(market_dict, title):
    fig = make_subplots(rows=1, cols=1)
    for key in market_dict.keys():
        df = market_dict[key]
        fig.add_trace(go.Scatter(x=df.index, y=df['Close'], mode='lines', name=key))
    fig.update_layout(title=title, xaxis_title='Date', yaxis_title='Close Price')
    fig.show()

plot_market_data(broad_market, 'Broad Market Assets Close Prices')
plot_market_data(developed_market, 'Developed Market Assets Close Prices')
plot_market_data(emerging_market, 'Emerging Market Assets Close Prices')
plot_market_data(frontier_market, 'Frontier Market Assets Close Prices')
```
