```python
major_pairs = md.get_forex_data()
minor_pairs = md.get_forex_data(category='minor pairs')
emerging_market_pairs = md.get_forex_data(category='emerging market pairs')
safe_haven_currencies = md.get_forex_data(category='safe-haven currencies')
major_currencies = md.get_forex_data(category='major currencies')
emerging_exotic_currencies = md.get_forex_data(category='emerging / exotic currencies')
```

```python
def plot_market_data(market_dict, title):
    fig = make_subplots(rows=1, cols=1)
    for key in market_dict.keys():
        df = market_dict[key]
        fig.add_trace(go.Scatter(x=df.index, y=df['Close'], mode='lines', name=key))
    fig.update_layout(title=title, xaxis_title='Date', yaxis_title='Close Price')
    fig.show()
```

```python
plot_market_data(major_pairs, 'Major Forex Pairs')
plot_market_data(minor_pairs, 'Minor Forex Pairs')
plot_market_data(emerging_market_pairs, 'Emerging Market Forex Pairs')
plot_market_data(safe_haven_currencies, 'Safe Haven Currencies')
plot_market_data(major_currencies, 'Major Currencies')
plot_market_data(emerging_exotic_currencies, 'Emerging / Exotic Currencies')
```
